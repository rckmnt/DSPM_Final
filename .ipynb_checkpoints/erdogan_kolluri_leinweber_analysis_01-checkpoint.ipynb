```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
import sys
reload(sys)
sys.setdefaultencoding('utf8')
```

```python
coke_sents = ['coke_ADJ_sent', 'coke_ADV_sent', 'coke_NOUN_sent', 'coke_VERB_sent']
pepsi_sents = ['pepsi_ADJ_sent', 'pepsi_ADV_sent', 'pepsi_NOUN_sent', 'pepsi_VERB_sent']

pos = ['ADJ', 'ADV', 'NOUN', 'VERB']

coke_hashes = 'coke_hashes.txt'
pepsi_hashes = 'pepsi_hashes.txt'

coke_pos = 'cokepos.csv'
pepsi_pos = 'pepsipos.csv'
```

## Sentiments

```python
cs_dfs = {c: pd.read_csv('{}'.format(c),header=None, names = ['word', 'score']) for c in coke_sents}
ps_dfs = {p: pd.read_csv('{}'.format(p),header=None, names = ['word', 'score']) for p in pepsi_sents}

# print(coke_sent_dfs['coke_ADJ_sent'])
```

```python
# Adjectives 


# for p in pos:

#     cokelabel = 
    
# Coke 
print("Coke's total Sentiment score for Adjectives:\n{}\n".format(cs_dfs['coke_ADJ_sent']['score'].sum()))
cdf = cs_dfs['coke_ADJ_sent']
cdf['company'] = 'coke'
print(cdf[cdf['score']  < 0 ])


# Pepsi
print("Pepsi's total Sentiment score for Adjectives:\n{}".format(ps_dfs['pepsi_ADJ_sent']['score'].sum()))
pdf = ps_dfs['pepsi_ADJ_sent']
pdf['company'] = 'pepsi'
print(pdf[pdf['score']  < 0 ])

# Combine
mdf = cdf.append(pdf, ignore_index=True)
mdf.sort_values('score', inplace=True)



# Plotting

plt.style.use('ggplot')
plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off') # labels along the bottom edge are off

colors = ['red' if _y =='coke' else 'blue' for _y in mdf['company']]
# colors = ['red', 'blue']
# ax.set_color_cycle(['red', 'blue'])

plt.legend(bbox_to_anchor=(1.05, 0), loc=2, borderaxespad=0.)

ax = mdf['score'].plot(kind='bar', 
                       title ="Coke and Pepsi Adjective Sentiments", 
                       color = colors, 
                       figsize=(15, 8), 
                       legend=True, 
                       fontsize=12)



```

```python

# Adverbs 

print("Coke's total Sentiment score for Adverbs:\n{}\n".format(cs_dfs['coke_ADV_sent'][1].sum()))
print("Pepsi's total Sentiment score for Adverbs:\n{}".format(ps_dfs['pepsi_ADV_sent'][1].sum()))
```

```python
# Verbs 


print("Coke's total Sentiment score for Verb:\n{}\n".format(cs_dfs['coke_VERB_sent'][1].sum()))
print("Pepsi's total Sentiment score for Verb:\n{}".format(ps_dfs['pepsi_VERB_sent'][1].sum()))
```
