```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
```

```python
coke_sents = ['coke_ADJ_sent', 'coke_ADV_sent', 'coke_NOUN_sent', 'coke_VERB_sent']
pepsi_sents = ['pepsi_ADJ_sent', 'pepsi_ADV_sent', 'pepsi_NOUN_sent', 'pepsi_VERB_sent']

pos = ['ADJ', 'ADV', 'NOUN', 'VERB']
posdic = {'ADJ':'Adjective', 'ADV':'Adverb', 'NOUN':'Noun', 'VERB':'Verb'}


coke_hashes = 'coke_hashes.txt'
pepsi_hashes = 'pepsi_hashes.txt'

coke_pos = 'cokepos.csv'
pepsi_pos = 'pepsipos.csv'
```

## Sentiments

```python
# Make dictionaries of all four POS sentiment CSV turned into DFs
cs_dfs = {c: pd.read_csv('{}'.format(c),header=None, names = ['word', 'score']) for c in coke_sents}
ps_dfs = {p: pd.read_csv('{}'.format(p),header=None, names = ['word', 'score']) for p in pepsi_sents}
```

```python

# for each POS print word sentiments and unique negative words

for p in pos:

    print("_" * 20)
    print("\t" + str(p))
    print("_" * 20)
    
    cokelabel = 'coke_' + p + '_sent'
    pepsilabel = 'pepsi_' + p + '_sent'

    # Coke 
    print("Coke's total Sentiment score for " + posdic[p] + "s:\n{}\n".format(cs_dfs['coke_ADJ_sent']['score'].sum()))
    cdf = cs_dfs[cokelabel]
    cdf['company'] = 'coke'
    print(cdf[cdf['score']  < 0 ])
    print('\n')


    # Pepsi
    print("Pepsi's total Sentiment score for " + posdic[p] + "s:\n{}".format(ps_dfs['pepsi_ADJ_sent']['score'].sum()))
    pdf = ps_dfs[pepsilabel]
    pdf['company'] = 'pepsi'
    print(pdf[pdf['score']  < 0 ])

    # Combine
    mdf = cdf.append(pdf, ignore_index=True)
    mdf.sort_values('score', inplace=True)



    # Plotting

    plt.style.use('ggplot')
    plt.tick_params(
        axis='both',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        bottom='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelbottom='off') # labels along the bottom edge are off

    colors = ['red' if _y =='coke' else 'blue' for _y in mdf['company']]
    plt.legend(bbox_to_anchor=(1.05, 0), loc=2, borderaxespad=0.)

    ax = mdf['score'].plot(kind='bar', 
                           title ="Coke and Pepsi " + posdic[p] + " Sentiments", 
                           color = colors, 
                           figsize=(15, 8), 
                           legend=True, 
                           fontsize=12)
    plt.show()
```
